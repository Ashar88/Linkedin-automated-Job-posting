```python
import gspread
import pandas as pd
import openpyxl

from pprint import pprint

# pprint(vars(object)) # important information
```

```python
gc = gspread.oauth()
sh = gc.open("TRP Jobs Roster Team 3")
```

```python
spreadsheet_data = gc.export(sh.id,gspread.utils.ExportFormat.EXCEL)

file_path = 'Roster_File.xlsx'

# Write the binary data to a file
with open(file_path, 'wb') as file:
    file.write(spreadsheet_data)
```

```python

FILE_NAME = "Roster_File.xlsx"
SHEET = ""
FULLNAME = "Ashar Saleem"

wb = openpyxl.load_workbook(FILE_NAME)
sheets = wb.sheetnames
sheets
```

```python
latestWeek = sheets[sheets.index('Dashboards') - 1 -1 ]
latestWeek
```

```python
SHEET = latestWeek

df = pd.read_excel(FILE_NAME, sheet_name = SHEET)
df.head()
```

```python
ws = wb[SHEET]
columns = list(df.columns)

for index, row in enumerate(ws.iter_rows()):
    for col, cell in enumerate(row):
        if cell.hyperlink:
            print(cell.hyperlink.target, cell.value)
            df.loc[index-1, f'col-{  columns[col] }'] = cell.hyperlink.target
```

```python
df
```

```python
mask = df['Recruiter'].apply(lambda x: str(x).lower() == FULLNAME.lower() or str(x).lower() == FULLNAME.lower().split()[0])

df = df[mask]
df
```
